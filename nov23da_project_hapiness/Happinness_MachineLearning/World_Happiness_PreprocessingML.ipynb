```python
import pandas as pd
import numpy as np

#Importer les fichiers
df_all = pd.read_csv("world-happiness-report.csv")
pd.set_option("display.max_columns", None) #Pour afficher toutes les colonnes

df_2021 = pd.read_csv("world-happiness-report-2021.csv")
pd.set_option("display.max_columns", None) #Pour afficher toutes les colonnes


#Visualiser les fichiers
display(df_all.head())
display(df_2021.head())

#Dimension 
print("Il y a", len(df_all["Country name"].unique()), "pays différents dans df_all")
print("Il y a", len(df_2021["Country name"].unique()), "pays différents dans df_2021")
df_all.isna().sum()*100/len(df_all)
```

```python
#Renomer les colonnes pour df_all
df_all.rename(columns={'Country name': 'Country', 
                    'Regional indicator': 'Region', 
                    'Life Ladder': 'Ladder_score',
                   "Log GDP per capita" : "PIB_habitant",
                   "year":"Year",
                   "Healthy life expectancy at birth" : "Healthy life expectancy",
                   "Freedom to make life choices" : "Freedom",
                   "Perceptions of corruption" : "Corruption"}, inplace=True)

df_all.columns = df_all.columns.str.replace(" ", "_")



#Renomer les colonnes pour df_all df_2021
df_2021.rename(columns={'Country name': 'Country', 
                    'Regional indicator': 'Region', 
                    'Ladder score': 'Ladder_score',
                   "Logged GDP per capita" : "PIB_habitant",
                   "Freedom to make life choices" : "Freedom",
                   "Perceptions of corruption" : "Corruption",
                   "Explained by: Log GDP per capita" : "Explained_by_PIB",
                   "Explained by: Social support" : "Explained by Social support",
                   "Explained by: Healthy life expectancy" : "Explained by Healthy life expectancy",
                   "Explained by: Freedom to make life choices" : "Explained by Freedom",
                   "Explained by: Generosity" : "Explained by Generosity",
                   "Explained by: Perceptions of corruption" : "Explained by corruption"}, inplace=True)

df_2021.columns = df_2021.columns.str.replace(" ", "_")


display(df_all.head())
display(df_2021.head())
```

```python
#Suppression des colonnes inutiles

df_2021 = df_2021.drop(["Standard_error_of_ladder_score", "upperwhisker", "lowerwhisker",
                       "Ladder_score_in_Dystopia", "Explained_by_PIB", "Explained_by_Social_support",
                        "Explained_by_Healthy_life_expectancy", "Explained_by_Freedom", "Explained_by_Generosity",
                        "Explained_by_corruption", "Dystopia_+_residual"], axis = 1)
```

```python
display(df_all.head())
display(df_2021.head())


print("Nous avons", df_all.Year.isna().sum(), "valeur(s) manquante(s) dans Year") #Pas de valeur manquante ds Year
```

```python
#Merge des 2 fichiers pour avoir les régions pour tous les pays

df_merge = pd.merge(df_all, df_2021, on = "Country", how = "outer")

df_merge.head()
```

```python
#Visualisation de la dimension des fichiers

print(df_all.shape)
print(df_merge.shape)

#On voit qu'il y a seulement 1949 lignes, il nous en manque donc des lignes
```

```python
#Concaténation de df_merge et df_2021

df_final = pd.concat([df_merge, df_2021], ignore_index=True, axis = 0)


#Tri pour vérifier que Afghanistan 2021 a bien été ajouté
df_final = df_final.sort_values(by = ["Country", "Year"]) 


print(df_final.shape) #Les lignes on bien étaient rajoutées mais trop de colonnes

df_final = df_final.reset_index(drop = True)

df_final.head(13) #Pour avoir un visuel sur le Afghanistan 2021
```

```python
#On remet les données de 2021 à la bonne place

df_final["Year"].fillna(2021, inplace = True) #Ajout de 2021 au NaN de Year 
df_final["Year"]= df_final["Year"].astype("int") #Year au format int


df_final["Ladder_score_x"].fillna(df_final["Ladder_score"], inplace= True)
df_final["PIB_habitant_x"].fillna(df_final["PIB_habitant"], inplace= True)
df_final["Social_support_x"].fillna(df_final["Social_support"], inplace= True)
df_final["Healthy_life_expectancy_x"].fillna(df_final["Healthy_life_expectancy"], inplace= True)
df_final["Freedom_x"].fillna(df_final["Freedom"], inplace= True)
df_final["Generosity_x"].fillna(df_final["Generosity"], inplace= True)
df_final["Corruption_x"].fillna(df_final["Corruption"], inplace= True)

df_final.head(13) #Pour avoir un visuel sur le Afghanistan 2021. Si on enlève .head(13), on a visuel sur Zimbabwe
```

```python
#Suppression des colonnes inutiles

df_final.drop(df_final.columns[-14:], axis = 1, inplace = True) #Suppression des 14 dernières colonnes
                                                            #A partir de Ladder_score_y

df_final.head()
```

```python
#On mets les noms des colonnes au propre

df_final.columns = df_final.columns.str.replace("_x", "")


df_final.head()
```

```python
#Taux de valeurs manquantes (Modifier chacun de son coté les NaN pour le ML)

(df_final.isna().sum())*100/len(df_final)
```

```python
#Suppression des Région NaN

df_final.dropna(subset = "Region", inplace = True)

(df_final.isna().sum())*100/len(df_final)
```

```python
df_final["Positive_affect"].fillna(method = "pad", inplace = True) #Par la valeur juste avant
df_final["Negative_affect"].fillna(method = "pad", inplace = True) #Par la valeur juste avant

df_final.fillna(method = "bfill", inplace = True) #Par la valeur juste après

print(df_final.isna().sum())
```

```python
# Supprimer la colonne 'Country'
df_final = df_final.drop(['Country', 'Year'], axis=1)

# Utiliser get_dummies pour encoder la colonne 'Region'
df_final = pd.get_dummies(df_final, columns=['Region'])

# Afficher le DataFrame résultant
df_final.head()
```
